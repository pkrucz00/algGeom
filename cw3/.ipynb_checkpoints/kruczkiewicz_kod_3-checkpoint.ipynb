### Konfiguracja

```python
# Narzędzie jest oparte o kilka zewnętrznych bibliotek, które potrzebujemy najpierw zaimportować.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as mcoll
import matplotlib.colors as mcolors
from matplotlib.widgets import Button
import json as js

# Parametr określający jak blisko (w odsetku całego widocznego zakresu) punktu początkowego 
# wielokąta musimy kliknąć, aby go zamknąć.
TOLERANCE = 0.15

def dist(point1, point2):
    return np.sqrt(np.power(point1[0] - point2[0], 2) + np.power(point1[1] - point2[1], 2))

# Klasa ta trzyma obecny stan wykresu oraz posiada metody, które mają zostać wykonane
# po naciśnięciu przycisków.
class _Button_callback(object):
    def __init__(self, scenes):
        self.i = 0
        self.scenes = scenes
        self.adding_points = False
        self.added_points = []
        self.adding_lines = False
        self.added_lines = []
        self.adding_rects = False
        self.added_rects = []

    def set_axes(self, ax):
        self.ax = ax
        
    # Metoda ta obsługuje logikę przejścia do następnej sceny.
    def next(self, event):
        self.i = (self.i + 1) % len(self.scenes)
        self.draw(autoscaling = True)

    # Metoda ta obsługuje logikę powrotu do poprzedniej sceny.
    def prev(self, event):
        self.i = (self.i - 1) % len(self.scenes)
        self.draw(autoscaling = True)
        
    # Metoda ta aktywuje funkcję rysowania punktów wyłączając równocześnie rysowanie 
    # odcinków i wielokątów.
    def add_point(self, event):
        self.adding_points = not self.adding_points
        self.new_line_point = None
        if self.adding_points:
            self.adding_lines = False
            self.adding_rects = False
            self.added_points.append(PointsCollection([]))
            
    # Metoda ta aktywuje funkcję rysowania odcinków wyłączając równocześnie
    # rysowanie punktów i wielokątów.     
    def add_line(self, event):   
        self.adding_lines = not self.adding_lines
        self.new_line_point = None
        if self.adding_lines:
            self.adding_points = False
            self.adding_rects = False
            self.added_lines.append(LinesCollection([]))

    # Metoda ta aktywuje funkcję rysowania wielokątów wyłączając równocześnie
    # rysowanie punktów i odcinków.
    def add_rect(self, event):
        self.adding_rects = not self.adding_rects
        self.new_line_point = None
        if self.adding_rects:
            self.adding_points = False
            self.adding_lines = False
            self.new_rect()
    
    def new_rect(self):
        self.added_rects.append(LinesCollection([]))
        self.rect_points = []
        
    # Metoda odpowiedzialna za właściwą logikę rysowania nowych elementów. W
    # zależności od włączonego trybu dodaje nowe punkty, początek, koniec odcinka
    # lub poszczególne wierzchołki wielokąta. Istnieje ciekawa logika sprawdzania
    # czy dany punkt jest domykający dla danego wielokąta. Polega ona na tym, że
    # sprawdzamy czy odległość nowego punktu od początkowego jest większa od
    # średniej długości zakresu pomnożonej razy parametr TOLERANCE.   
    def on_click(self, event):
        if event.inaxes != self.ax:
            return
        new_point = (event.xdata, event.ydata)
        if self.adding_points:
            self.added_points[-1].add_points([new_point])
            self.draw(autoscaling = False)
        elif self.adding_lines:
            if self.new_line_point is not None:
                self.added_lines[-1].add([self.new_line_point, new_point])
                self.new_line_point = None
                self.draw(autoscaling = False)
            else:
                self.new_line_point = new_point
        elif self.adding_rects:
            if len(self.rect_points) == 0:
                self.rect_points.append(new_point)
            elif len(self.rect_points) == 1:
                self.added_rects[-1].add([self.rect_points[-1], new_point])
                self.rect_points.append(new_point)
                self.draw(autoscaling = False)
            elif len(self.rect_points) > 1:
                if dist(self.rect_points[0], new_point) < (np.mean([self.ax.get_xlim(), self.ax.get_ylim()])*TOLERANCE):
                    self.added_rects[-1].add([self.rect_points[-1], self.rect_points[0]])
                    self.new_rect()
                else:    
                    self.added_rects[-1].add([self.rect_points[-1], new_point])
                    self.rect_points.append(new_point)
                self.draw(autoscaling = False)
    
    # Metoda odpowiedzialna za narysowanie całego wykresu. Warto zauważyć,
    # że zaczyna się ona od wyczyszczenia jego wcześniejszego stanu. Istnieje w
    # niej nietrywialna logika zarządzania zakresem wykresu, tak żeby, w zależności
    # od ustawionego parametru autoscaling, uniknąć sytuacji, kiedy dodawanie
    # nowych punktów przy brzegu obecnie widzianego zakresu powoduje niekorzystne
    # przeskalowanie.
    def draw(self, autoscaling = True):
        if not autoscaling:
            xlim = self.ax.get_xlim()
            ylim = self.ax.get_ylim()
        self.ax.clear()
        for collection in (self.scenes[self.i].points + self.added_points):
            if len(collection.points) > 0:
                self.ax.scatter(*zip(*(np.array(collection.points))), **collection.kwargs)
        for collection in (self.scenes[self.i].lines + self.added_lines + self.added_rects):
            self.ax.add_collection(collection.get_collection())
        self.ax.autoscale(autoscaling)
        if not autoscaling:
            self.ax.set_xlim(xlim)
            self.ax.set_ylim(ylim)
        plt.draw()
```

```python
# Klasa Scene odpowiada za przechowywanie elementów, które mają być
# wyświetlane równocześnie. Konkretnie jest to lista PointsCollection i
# LinesCollection.
class Scene:
    def __init__(self, points=[], lines=[]):
        self.points=points
        self.lines=lines

# Klasa PointsCollection gromadzi w sobie punkty jednego typu, a więc takie,
# które zostaną narysowane w takim samym kolorze i stylu. W konstruktorze
# przyjmuje listę punktów rozumianych jako pary współrzędnych (x, y). Parametr
# kwargs jest przekazywany do wywołania funkcji z biblioteki MatPlotLib przez
# co użytkownik może podawać wszystkie parametry tam zaproponowane.        
class PointsCollection:
    def __init__(self, points, **kwargs):
        self.points = points
        self.kwargs = kwargs
    
    def add_points(self, points):
        self.points = self.points + points

# Klasa LinesCollection podobnie jak jej punktowy odpowiednik gromadzi
# odcinki tego samego typu. Tworząc ją należy podać listę linii, gdzie każda
# z nich jest dwuelementową listą punktów – par (x, y). Parametr kwargs jest
# przekazywany do wywołania funkcji z biblioteki MatPlotLib przez co użytkownik
# może podawać wszystkie parametry tam zaproponowane.
class LinesCollection:
    def __init__(self, lines, **kwargs):
        self.lines = lines
        self.kwargs = kwargs
        
    def add(self, line):
        self.lines.append(line)
        
    def get_collection(self):
        return mcoll.LineCollection(self.lines, **self.kwargs)

# Klasa Plot jest najważniejszą klasą w całym programie, ponieważ agreguje
# wszystkie przygotowane sceny, odpowiada za stworzenie wykresu i przechowuje
# referencje na przyciski, dzięki czemu nie będą one skasowane podczas tzw.
# garbage collectingu.
class Plot:
    def __init__(self, scenes = [Scene()], points = [], lines = [], json = None):
        if json is None:
            self.scenes = scenes
            if points or lines:
                self.scenes[0].points = points
                self.scenes[0].lines = lines
        else:
            self.scenes = [Scene([PointsCollection(pointsCol) for pointsCol in scene["points"]], 
                                 [LinesCollection(linesCol) for linesCol in scene["lines"]]) 
                           for scene in js.loads(json)]
    
    # Ta metoda ma szczególne znaczenie, ponieważ konfiguruje przyciski i
    # wykonuje tym samym dość skomplikowaną logikę. Zauważmy, że konfigurując każdy
    # przycisk podajemy referencję na metodę obiektu _Button_callback, która
    # zostanie wykonana w momencie naciśnięcia.
    def __configure_buttons(self):
        plt.subplots_adjust(bottom=0.2)
        ax_prev = plt.axes([0.6, 0.05, 0.15, 0.075])
        ax_next = plt.axes([0.76, 0.05, 0.15, 0.075])
        ax_add_point = plt.axes([0.44, 0.05, 0.15, 0.075])
        ax_add_line = plt.axes([0.28, 0.05, 0.15, 0.075])
        ax_add_rect = plt.axes([0.12, 0.05, 0.15, 0.075])
        b_next = Button(ax_next, 'Następny')
        b_next.on_clicked(self.callback.next)
        b_prev = Button(ax_prev, 'Poprzedni')
        b_prev.on_clicked(self.callback.prev)
        b_add_point = Button(ax_add_point, 'Dodaj punkt')
        b_add_point.on_clicked(self.callback.add_point)
        b_add_line = Button(ax_add_line, 'Dodaj linię')
        b_add_line.on_clicked(self.callback.add_line)
        b_add_rect = Button(ax_add_rect, 'Dodaj figurę')
        b_add_rect.on_clicked(self.callback.add_rect)
        return [b_prev, b_next, b_add_point, b_add_line, b_add_rect]
    
    def add_scene(self, scene):
        self.scenes.append(scene)
    
    def add_scenes(self, scenes):
        self.scenes = self.scenes + scenes

    # Metoda toJson() odpowiada za zapisanie stanu obiektu do ciągu znaków w
    # formacie JSON.
    def toJson(self):
        return js.dumps([{"points": [np.array(pointCol.points).tolist() for pointCol in scene.points], 
                          "lines":[linesCol.lines for linesCol in scene.lines]} 
                         for scene in self.scenes])    
    
    # Metoda ta zwraca punkty dodane w trakcie rysowania.
    def get_added_points(self):
        if self.callback:
            return self.callback.added_points
        else:
            return None
    
    # Metoda ta zwraca odcinki dodane w trakcie rysowania.
    def get_added_lines(self):
        if self.callback:
            return self.callback.added_lines
        else:
            return None
        
    # Metoda ta zwraca wielokąty dodane w trakcie rysowania.
    def get_added_figure(self):
        if self.callback:
            return self.callback.added_rects
        else:
            return None
    
    # Metoda ta zwraca punkty, odcinki i wielokąty dodane w trakcie rysowania
    # jako scenę.
    def get_added_elements(self):
        if self.callback:
            return Scene(self.callback.added_points, self.callback.added_lines+self.callback.added_rects)
        else:
            return None
    
    # Główna metoda inicjalizująca wyświetlanie wykresu.
    def draw(self):
        plt.close()
        fig = plt.figure()
        self.callback = _Button_callback(self.scenes)
        self.widgets = self.__configure_buttons()
        ax = plt.axes(autoscale_on = False)
        self.callback.set_axes(ax)
        fig.canvas.mpl_connect('button_press_event', self.callback.on_click)
        plt.show()
        self.callback.draw()
        
```

### Punkt 2 i 3 - y-monotoniczność i podział wierzchołków
#### Wyznacznik i podział punktów
Użyto wyznacznika 3x3, który najlepiej sprawdził się w sprawozdaniu 1

```python
def det3x3(a,b,c):
    s1 = a[0]*b[1] + b[0]*c[1] + c[0]*a[1]
    s2 = a[1]*b[0] + b[1]*c[0] + c[1]*a[0]
    return s1 - s2

def divide(a,b,c):
    eps = 10**(-13)

    det = det3x3(a, b, c)   # computing the determinant of given interval and the point c
    if det < -eps:           # c is clockwise to the interval - it's on its right 
        return -1
    elif det > eps:          # c is counterclockwise to the interval - it's on its left
        return 1
    else:                    # c is within the "epsilon range" - collinear
        return 0
```

#### Podział na punkty początkowe, końcowe, łączące, dzielące i prawidłowe
Funkcja `splitPoints` zwraca słownik, którego kluczami są nazwy grup, a wartością - lista wierzchołków znajdujących się w danej grupie

```python
def splitPoints(P):
    def getNeighboursIndexes(i,n):
        return (i-1)%n, (i+1)%n
    
    def isAbove(p, q):  #returns True if p is above q
        return p[1] > q[1]
    
    n = len(P)
    result = {}
    
    for i in range(n):
        l, r = getNeighboursIndexes(i,n)
        current, left, right = P[i], P[l], P[r]
        div = divide(left, current, right)
        
        if div == 0:
            result[current] = "regular"
        elif isAbove(current, left) and isAbove(current,right):    
            if div > 0:
                result[current] = "start"
            elif div < 0:
                result[current] = "split"
        elif isAbove(left, current) and isAbove(right, current):
            if div > 0:
                result[current] = "end"
            elif div < 0:
                result[current] = "merge"
        else:
            result[current] = "regular"
            
    return result
```

### Sprawdzenie y-monotonicznosci
Korzystamy z faktu, że jeżeli wielokąt jest y-monotoniczny, to nie posiada wierzchołków dzielących oraz łączących

*Test* - sprawdzamy monotoniczność narysowanych wyżej punktów (`True` jeżeli jest Y-monotoniczna i `False` wpp)

```python
def isYmonotone(figure, divOfPoints = None):
    if divOfPoints is None: 
        divOfPoints = splitPoints(figure)
    return not ("split" in divOfPoints.values() or "merge" in divOfPoints.values())

#Test
figures = getFigurePoints(plot1)
for figure in figures:
    print(isYmonotone(figure))
```

### punkt 4 - triangulacja
#### Wizualizator
Wizualizator to klasa zachowująca kolory, stan działania algorytmu (w atrybucie `results`), obwód (`circumference`) oraz linie - przekątne "rozdzielające" trjókąty w wielokącie (`triangleDiagonals`).

Obiekt tej klasy dostaje w konstruktorze listy stanu. W metodzie `makeScene()` dodaje obiekt `Scene` do atrybutu `scenes` i na koniec zwraca te sceny metodą `getScenes()`.

```python
class Visualizer:
    def __init__(self, arrays):
        self.__colors = ["blue", "green", "red", "gold"]
        self.circumference = self._makeLines(arrays[0])
        self.scenes = [Scene([PointsCollection(arrays[0][:], color=self.__colors[0])], 
                             [LinesCollection(self.circumference, color=self.__colors[0])])]
        self.results = arrays     #array[0] - default points, array[1] - already processed,
                                  #array[2] - currently processed, array[3] - on stack
        self.triangleDiagonals = []
        
    
    def _makeLines(self, pointSet):
        if len(pointSet) < 2:  return [] 
        
        n = len(pointSet)
        lines = [[pointSet[i%n], pointSet[(i+1)%n]] for i in range(n)]

        return lines

    
    def makeScene(self):
        points = [PointsCollection(self.results[i][:], color=self.__colors[i]) for i in range(len(self.results))]
        lines = [LinesCollection(self.circumference, color=self.__colors[0]),
                 LinesCollection(self.triangleDiagonals[:], color=self.__colors[1])]
        self.scenes.append(Scene(points, lines))
        
    
    def getScenes(self):
        return self.scenes
        
```

#### Przygotowanie struktury
Zakładamy, że dostajemy zbiór punktów tworzących wielokąt, które są już posortowane przeciwnie do ruchu wskazówek zegara.

Funkcja `splitIntoChains` znajduje indeks najwyższego i najniższego elementu i rozdziela wierzchołki na znajdujące się po lewej i prawej stronie. Łańcuch po drugiej stronie zostaje również odwrócony, aby punkty były uporządkowane w kolejności malejącej względem współrzędnej y 

Funkcja `sortAndPreparePoints` bierze dwa łańcuchy - lewy i prawy - i analogicznie do funkcji `merge` w mergeSorcie, sortuje je w jeden łańcuch. Wynikowa lista składa się jednak z tupli w formie `(punkt, nazwa łańcucha)`. Wiadomość, w którym łańcuchu znajduje się dany punkt jest ważna dla działania algorytmu, więc została dodana w tuplu.

```python
def splitIntoChains(figure):
    high = max(list(range(len(figure))), key=lambda i: figure[i][1])
    low = min(list(range(len(figure))), key=lambda i: figure[i][1])
    figure = figure[high:] + figure[:high]  #we reconfigure the point so that the highest is first
    high, low = 0, low-high  #computing new indexes of lowest and highest elemnts
    return figure[high], figure[low], figure[high+1:low], figure[low+1:][::-1]  #the last one is reversed so the points are sorted in descending order with regards to y coordinate


def sortAndPreparePoints(leftChain, rightChain):
    result = []
    n1, n2 = len(leftChain), len(rightChain)
    i, j = 0, 0
    
    while i < n1 and j < n2:
        pointLeft, pointRight = leftChain[i], rightChain[j]
        if pointLeft[1] >= pointRight[1]:
            result.append((pointLeft, "left"))
            i += 1
        else:
            result.append((pointRight, "right"))
            j += 1
    
    while i < n1:
        result.append((leftChain[i], "left"))
        i += 1
    while j < n2:
        result.append((rightChain[j], "right"))
        j += 1
    
    return result
```

#### Funkcja główna
Bierzemy wielokąt przechowywanych w formie wierzchołków, a zwracamy listę przekątnych wielokąta wchodzących w skład triangulacji danego wielokąta.
Zakładamy również, że wielokąt jest ścieśle y-monotoniczny.

Funkcja korzysta z wyżej opisanych dwóch funkcji, aby przygotować dane.
Funkcja pomocnicza `isInFigure` sprawdza, czy dany trójkąt znajduje się wewnątrz wielokąta.

*Test* - sprawdzamy triangulację pierwszego z narysowanych wyżej wielokątów. Zwracamy tablicę zawierającą linie. Uwaga - jeżeli narysowany wielokąt nie jest Y-monotoniczny, dostaniemy `assertionError`

```python
def triangulate(figure):
    assert isYmonotone(figure), "Figure is not Y-monotone"
    
    def isInFigure(a, b, c, chain):    
        if chain == "left":
            return divide(a, b, c) > 0
        else:
            return divide(a, b, c) < 0

    pHigh, pLow, leftChain, rightChain = splitIntoChains(figure)
    preparedArr = sortAndPreparePoints(leftChain, rightChain)

    result = []
    stack = [(pHigh, None), preparedArr[0]]    #first element is (pHigh, None) because highest point is nither on left nor right chain.
    for i, (p, chain) in enumerate(preparedArr[1:], start=0):   #enumerate function provides index of the processed element
        v, v_chain = stack[-1]
        if chain != v_chain:    #if point is on the chain opposite to last stack element 
            while len(stack) > 1:
                v, v_chain = stack.pop()
                result.append((v, p))
            stack = [preparedArr[i], (p, chain)]  #stack consists of processed element and te one proceeding it (preparedArr[i])
        else:
            v, v_chain = stack.pop()
            u = None
            while len(stack) > 0 and isInFigure(stack[-1][0], v, p, chain):    #while the triangle is in figure, we add points
                u, u_chain = stack.pop()
                result.append((u, p))
            if u is None: stack.append((v, v_chain))   #the last popped element goes back on stack
            else: stack.append((u, u_chain))

            stack.append((p, chain))    #we add point to the stack

    stack.pop()
    while len(stack) > 1:           #all the elements from stack besides the last and the first make a valid triangle with the lowest element
        u, u_chain = stack.pop()
        result.append((u, pLow))

    return result

figures = getFigurePoints(plot1)
triangulation = print(len(triangulate(figures[0])))
```

#### Wersja z wizualizacją
*Test* - wizualizujemy triangulację pierwszego wielokąta z narysowanych wyżej 

```python
def triangulateShow(figure):
    assert isYmonotone(figure), "Figure is not Y-monotone"
    
    def isInFigure(a, b, c, chain):
        if chain == "left":
            return divide(a, b, c) > 0
        else:
            return divide(a, b, c) < 0
      
    
    state = [[] for _ in range(4)]
    state[0] = figure
    vis = Visualizer(state)
    
    pHigh, pLow, leftChain, rightChain = splitIntoChains(figure)
    preparedArr = sortAndPreparePoints(leftChain, rightChain)

    result = []
    stack = [(pHigh, None), preparedArr[0]]
    
    
    state[3] = list(map(lambda x: x[0], stack))
    state[1] = state[3][:]
    
    vis.triangleDiagonals = result
    vis.makeScene()
    
    for i, (p, chain) in enumerate(preparedArr[1:]):
        state[2] = [p]
        v, v_chain = stack[-1]
        if chain != v_chain:
            while len(stack) > 1:
                state[3] = list(map(lambda x: x[0], stack))
                vis.makeScene()
                
                v, v_chain = stack.pop()
                result.append((v, p))
                
            stack = [preparedArr[i], (p, chain)]
            
            state[1].append(p)
            state[3] = list(map(lambda x: x[0], stack))
            vis.makeScene()
        else:
            state[3] = list(map(lambda x: x[0], stack))
            vis.makeScene()
            
            v, v_chain = stack.pop()
            u = None
            while len(stack) > 0 and isInFigure(stack[-1][0], v, p, chain):
                state[3] = list(map(lambda x: x[0], stack))
                vis.makeScene()
                
                u, u_chain = stack.pop()
                result.append((u, p))
                v, v_chain = u, u_chain
            
            if u is None: stack.append((v, v_chain))
            else: 
                stack.append((u, u_chain))
                
            stack.append((p, chain))
            state[1].append(p)
            
            state[3] = list(map(lambda x: x[0], stack))
            vis.makeScene()

    stack.pop()
    while len(stack) > 1:
        state[3] = list(map(lambda x: x[0], stack))
        
        u, u_chain = stack.pop()
        result.append((u, pLow))
        vis.makeScene()
        
    state[0] = state[2] = state[3] = []
    state[1] = figure
    vis.makeScene()

    return Plot(scenes=vis.getScenes())
```

```python
figures = getFigurePoints(plot1)
print(figures[0])
plot = triangulateShow(figures[0])
plot.draw()
```
